```python
from pathlib import Path

import pandas as pd

base_dir = Path("../results/ESZ4_ESH5_20241212/scalein_experiment")
trade_paths = {
    "European_Baseline": base_dir / "European_Baseline" / "trades.parquet",
    "European_V1": base_dir / "European_V1" / "trades.parquet",
    "US_RTH_Baseline": base_dir / "US_RTH_Baseline" / "trades.parquet",
    "US_RTH_V1": base_dir / "US_RTH_V1" / "trades.parquet",
    "Post_close_Baseline": base_dir / "Post_close_Baseline" / "trades.parquet",
    "Post_close_V1": base_dir / "Post_close_V1" / "trades.parquet",
}

trade_frames = {name: pd.read_parquet(path) for name, path in trade_paths.items()}
trade_summary = pd.DataFrame(
    {
        "strategy": list(trade_frames.keys()),
        "trades": [len(frame) for frame in trade_frames.values()],
        "path": [str(path) for path in trade_paths.values()],
    }
)

trade_summary
```
